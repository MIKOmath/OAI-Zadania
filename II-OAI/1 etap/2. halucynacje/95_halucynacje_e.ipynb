## Wstęp

Modele językowe pomagają nam w codziennych zadaniach, takich jak poprawianie tekstów, pisanie kodu czy odpowiadanie na pytania. 
Są one również coraz częściej wykorzystywane w takich dziedzinach jak medycyna czy edukacja.

Jednak skąd możemy wiedzieć, czy wygenerowane przez nie odpowiedzi są poprawne? Modele językowe nie zawsze posiadają pełną wiedzę na zadany temat, a mimo to mogą formułować odpowiedzi, które brzmią wiarygodnie, lecz w rzeczywistości wprowadzają w błąd. Takie niepoprawne odpowiedzi nazywamy halucynacjami.

### Kryterium Oceny

ROC AUC (ang. *Receiver Operating Characteristic Area Under Curve*) to miara jakości klasyfikatora binarnego. Pokazuje zdolność modelu do odróżniania między dwiema klasami - tutaj halucynacją (false) i poprawną odpowiedzią (true). 

- **ROC (Receiver Operating Characteristic)**: Wykres pokazujący zależność między *True Positive Rate* (czułość) a *False Positive Rate* (1-specyficzność) przy różnych progach decyzyjnych.
- **AUC (Area Under Curve)**: Pole pod wykresem ROC, które przyjmuje wartości od 0 do 1:
  - **1.0**: Model perfekcyjny.
  - **0.5**: Model losowy (brak zdolności do odróżniania klas).

Im wyższa wartość AUC, tym lepiej model radzi sobie z klasyfikacją.

Za to zadanie możesz zdobyć pomiędzy 0 a 100 punktów. Wynik będzie skalowany liniowo w zależności od wartości ROC AUC:

- **ROC AUC ≤ 0.7**: 0 punktów.
- **ROC AUC ≥ 0.82**: 100 punktów.
- **Wartości pomiędzy 0.7 a 0.82**: skalowane liniowo.

Wzór na wynik:  
$$
\text{Punkty} = 
\begin{cases} 
0 & \text{dla } \text{ROC AUC} \leq 0.7 \\
100 \times \frac{\text{ROC AUC} - 0.7}{0.82 - 0.7} & \text{dla } 0.7 < \text{ROC AUC} < 0.82 \\
100 & \text{dla } \text{ROC AUC} \geq 0.82
\end{cases}
$$

## Kod z Kryterium Oceniającym

Kod, zbliżony do poniższego, będzie używany do oceny rozwiązania na zbiorze testowym.

````python
######################### NIE ZMIENIAJ TEJ KOMÓRKI PODCZAS WYSYŁANIA ##########################

def compute_score(roc_auc: float) -> float:
    """
    Oblicza wynik punktowy na podstawie wartości ROC AUC.

    :param roc_auc: Wartość float w zakresie [0.0, 1.0]
    :return: Wynik punktowy zgodny z określoną funkcją
    """
    if roc_auc <= 0.7:
        return 0
    elif 0.7 < roc_auc < 0.82:
        return int(round(100 * (roc_auc - 0.7) / (0.82 - 0.7)))
    else:
        return 100


def evaluate_algorithm(dataset, algorithm, verbose=False):
    """
    Ewaluacja algorytmu wykrywania halucynacji na podanym zbiorze danych.

    Parametry
    ----------
    dataset : list
        Oznaczony zbiór danych, gdzie każdy element to słownik zawierający klucz 'is_correct'.
    algorithm : callable
        Funkcja, która przyjmuje pojedynczy przykład (słownik) i zwraca prawdopodobieństwo halucynacji.
    verbose : bool
        Jeśli True, wypisuje dodatkowe informacje dla każdego przykładu oraz podsumowanie.

    Zwraca
    -------
    roc_auc : float
        Wartość pola pod krzywą ROC (ROC AUC) dla predykcji.
    """
    predicted_ys = [] # Lista przechowująca przewidywane prawdopodobieństwa halucynacji

    for i, entry in enumerate(dataset):
        # Tworzenie kopii próbki i usunięcie etykiety, aby uzyskać dane wejściowe bez oznaczeń
        sample_unlabeled = dict(entry)
        sample_unlabeled.pop('is_correct', None)

        try:
            # Przewidywanie prawdopodobieństwa dla pojedynczej próbki
            pred_prob = algorithm(sample_unlabeled)
            predicted_ys.append(pred_prob)

        except Exception as e:
            # Jeśli wystąpi błąd, domyślnie ustawiamy prawdopodobieństwo na 0.5
            predicted_ys.append(0.5)
            if verbose:
                print(f"Sample {i} => Error: {e}")

    predicted_ys = np.array(predicted_ys, dtype=np.float32)
    ys = []
    for entry in dataset:
        ys.append(1 if entry.get('is_correct') else 0)
    ys = np.array(ys, dtype=np.int32)
    
    # Obliczenie metryki ROC AUC
    roc_auc = roc_auc_score(ys, predicted_ys)

    # Obliczenie końcowego wyniku na podstawie ROC AUC
    points = compute_score(roc_auc)

    if verbose:
        print(f"\nLiczba próbek: {len(dataset)}")
        print(f"ROC AUC: {roc_auc:.4f}")
        print(f"Wynik punktowy: {points}")

    return points
````

# Twoje Rozwiązanie
W tej sekcji należy umieścić Twoje rozwiązanie. Wprowadzaj zmiany wyłącznie tutaj!

````python
import re
import string
from collections import Counter

def safe_division(numerator, denominator, default_value=0.0):
    return default_value if denominator == 0 or np.isnan(numerator) or np.isnan(denominator) else numerator / denominator

def extract_answer_snippets(supporting_answers):
    pattern = re.compile(r'<answer>(.*?)</answer>')
    answer_snippets = [snippet for sa in supporting_answers for snippet in pattern.findall(sa)]
    return answer_snippets or supporting_answers

def create_features_combined(data):
    rows = []
    for rec in data:
        features = {}
        answer = rec.get('answer', '').lower()
        supporting_answers = rec.get('supporting_answers', [])

        all_probs = []
        for probs in rec.get('supporting_probabilities', []):
            all_probs.extend(probs)

        if all_probs:
            all_probs_arr = np.array(all_probs)
            features['prob_median'] = np.median(all_probs_arr)
            features['prob_range'] = np.max(all_probs_arr) - np.min(all_probs_arr)
            features['prob_std'] = np.std(all_probs_arr)

            if len(all_probs_arr) > 2:
                mean = np.mean(all_probs_arr)
                dev = all_probs_arr - mean
                m3 = np.mean(dev ** 3)
                m2 = np.mean(dev ** 2)
                features['prob_skew'] = m3 / (m2 ** 1.5) if m2 > 0 else 0.0

                m4 = np.mean(dev ** 4)
                features['prob_kurtosis'] = (m4 / (m2 ** 2)) - 3 if m2 > 0 else 0.0
            else:
                features['prob_skew'] = 0.0
                features['prob_kurtosis'] = 0.0

            features['low_prob_count'] = np.sum(all_probs_arr < 0.33)
            features['prob_q50'] = np.percentile(all_probs_arr, 50)
            q25, q75 = np.percentile(all_probs_arr, [25, 75])
            features['prob_iqr'] = q75 - q25
            features['avg_probability'] = np.mean(all_probs_arr)
        else:
            default = 0.5
            features['prob_median'] = default
            features['prob_range'] = 0.0
            features['prob_std'] = 0.1
            features['prob_skew'] = 0
            features['prob_kurtosis'] = 0.0
            features['low_prob_count'] = 0
            features['prob_q50'] = default
            features['prob_iqr'] = 0.1
            features['avg_probability'] = default

        answer_tokens = answer.split()
        answer_token_set = set(answer_tokens)

        if supporting_answers:
            answer_snippets = extract_answer_snippets(supporting_answers)
            confirmation_scores = [safe_division(len(set(snippet.lower().split()).intersection(answer_token_set)),
                                               max(len(snippet.lower().split()), 1)) for snippet in answer_snippets]
            features['confirmation_rate'] = safe_division(sum(1 for score in confirmation_scores if score > 0.7),
                                                      len(supporting_answers))

            clean_supporting_answers = [re.search(r'<answer>(.*?)</answer>', sa).group(1).strip()
                                      if re.search(r'<answer>(.*?)</answer>', sa) else sa
                                      for sa in supporting_answers]

            extract_numbers = lambda text: re.findall(r'\b\d+(?:\.\d+)?\b', text)
            main_numbers = extract_numbers(answer)
            supporting_numbers = [extract_numbers(sa) for sa in clean_supporting_answers]
            all_supporting_numbers = [num for sublist in supporting_numbers for num in sublist]

            features['unique_supporting_numbers'] = len(set(all_supporting_numbers))
            features['main_numbers_count'] = len(main_numbers)

            if main_numbers and all_supporting_numbers:
                overlap_count = sum(1 for num in main_numbers if num in all_supporting_numbers)
                features['number_overlap_ratio'] = safe_division(overlap_count, len(main_numbers))
                features['missing_numbers_ratio'] = safe_division(len(main_numbers) - overlap_count, len(main_numbers))
            else:
                features['number_overlap_ratio'] = 0 if main_numbers else 1
                features['missing_numbers_ratio'] = 1 if main_numbers else 0

            number_counter = Counter(all_supporting_numbers)
            most_common_numbers = number_counter.most_common(3)
            features['common_number_in_answer'] = sum((3 - i) / 6 for i, (num, count) in
                                                   enumerate(most_common_numbers[:min(3, len(most_common_numbers))])
                                                   if num in main_numbers)

            answer_words = set(word.strip(string.punctuation).lower()
                             for word in answer.split() if word.strip(string.punctuation))
            supporting_words_sets = [set(word.strip(string.punctuation).lower()
                                      for word in sa.split() if word.strip(string.punctuation))
                                    for sa in clean_supporting_answers]

            if answer_words:
                word_overlaps = [safe_division(len(answer_words.intersection(sw)), len(answer_words))
                               for sw in supporting_words_sets if sw]
                features['min_word_overlap'] = min(word_overlaps) if word_overlaps else 0
                features['avg_word_overlap'] = np.mean(word_overlaps) if word_overlaps else 0
            else:
                features['min_word_overlap'] = 0
                features['avg_word_overlap'] = 0

            get_bigrams = lambda text: set(zip(*(text.split()[i:] for i in range(2)))) if len(text.split()) > 1 else set()
            answer_bigrams = get_bigrams(answer.strip(string.punctuation).lower())
            supporting_bigrams_sets = [get_bigrams(sa.strip(string.punctuation).lower()) for sa in clean_supporting_answers]

            if answer_bigrams:
                bigram_overlaps = [safe_division(len(answer_bigrams.intersection(sb)), len(answer_bigrams))
                                 for sb in supporting_bigrams_sets if sb]
                features['min_bigram_overlap'] = min(bigram_overlaps) if bigram_overlaps else 0
            else:
                features['min_bigram_overlap'] = 0

            features['sentence_count'] = len([s for s in re.split(r'[.!?]+', answer) if s.strip()])
        else:
            features['confirmation_rate'] = 0.0
            features['missing_numbers_ratio'] = 1.0
            features['number_overlap_ratio'] = 0.0
            features['min_word_overlap'] = 0.0
            features['avg_word_overlap'] = 0.0
            features['unique_supporting_numbers'] = 0
            features['main_numbers_count'] = 0
            features['min_bigram_overlap'] = 0.0
            features['sentence_count'] = 0
            features['common_number_in_answer'] = 0.0

        features['unique_token_ratio'] = safe_division(len(set(answer_tokens)), len(answer_tokens))
        features['confirmation_prob_interaction'] = safe_division(features.get('confirmation_rate', 0) + 0.1, 1)
        prob_std = features['prob_std']
        prob_skew = features['prob_skew']
        iqr = features['prob_iqr']
        prob_mean = features['avg_probability']
        prob_median = features['prob_median']
        prob_kurtosis = features['prob_kurtosis']
        low_prob_count = features['low_prob_count']
        confirmation_rate = features['confirmation_rate']

        features['variability_score'] = prob_std * (abs(prob_skew) + 0.01) * (iqr + 0.01)
        features['confidence_score'] = prob_mean * prob_median * (abs(prob_kurtosis) + 0.01)
        features['hallucination_ratio'] = safe_division(features['variability_score'], features['confidence_score'] + 1e-10)
        features['low_prob_confirmation'] = low_prob_count * (1 - confirmation_rate + 0.01)

        features['is_correct'] = int(rec.get("is_correct", False))
        features['question_id'] = rec.get("question_id")

        supp_tokens_list = rec.get("supporting_tokens", [])
        supp_probs_list = rec.get("supporting_probabilities", [])

        answer_inside_mean = np.nan
        answer_inside_range = np.nan
        left_context_mean = np.nan

        if supp_tokens_list and supp_probs_list:
            supp_tokens = supp_tokens_list[0]
            supp_probs = supp_probs_list[0]

            probs = np.array(supp_probs)
            try:
                start_idx = next(i for i, t in enumerate(supp_tokens) if "<answer>" in t)
                end_idx = next(i for i, t in enumerate(supp_tokens[start_idx:]) if "</answer>" in t) + start_idx
                if end_idx - start_idx > 1:
                    inside_probs = probs[start_idx+1:end_idx]
                    answer_inside_mean = np.mean(inside_probs)
                    answer_inside_range = np.max(inside_probs) - np.min(inside_probs)
                if start_idx >= 3:
                    left_context_mean = np.mean(probs[start_idx-3:start_idx])
                elif start_idx > 0:
                    left_context_mean = np.mean(probs[:start_idx])
            except StopIteration:
                pass

        features['answer_inside_mean'] = answer_inside_mean
        features['answer_inside_range'] = answer_inside_range
        features['left_context_mean'] = left_context_mean

        rows.append(features)
    return rows

def preprocessing_func(X):
    return np.nan_to_num(X, nan=-1)

FEATURE_NAMES = [
    'low_prob_confirmation', 'answer_inside_mean', 'answer_inside_range', 'prob_median', 'hallucination_ratio',
    'confirmation_prob_interaction', 'unique_token_ratio', 'left_context_mean', 'prob_range', 'confidence_score',
    'missing_numbers_ratio', 'number_overlap_ratio', 'min_word_overlap', 'avg_probability', 'avg_word_overlap',
    'unique_supporting_numbers', 'main_numbers_count', 'min_bigram_overlap', 'prob_q50', 'sentence_count',
    'prob_iqr', 'common_number_in_answer'
]

MODEL_CONFIG = {
    'params': {
        'colsample_bytree': 0.7, 'gamma': 0.0, 'learning_rate': 0.01,
        'max_depth': 3, 'min_child_weight': 8, 'n_estimators': 600,
        'subsample': 0.73, 'objective': 'binary:logistic', 'eval_metric': 'logloss',
        'n_jobs': -1,
    },
}


def run_training(config, df_features):
    X = np.array([[rec.get(col, -1) for col in FEATURE_NAMES] for rec in df_features], dtype=float)
    y = np.array([rec["is_correct"] for rec in df_features], dtype=int)
    X_processed = preprocessing_func(X)
    
    model = xgb.XGBClassifier(**config['params'])
    model.fit(X_processed, y)
    
    return model

df_features_combined_train = create_features_combined(train)
final_model_result = run_training(MODEL_CONFIG, df_features_combined_train)
trained_model = final_model_result

def predict_hallucinations(sample):
    sample_features = create_features_combined([sample])
    X_sample = np.array([[rec.get(col, -1) for col in FEATURE_NAMES] for rec in sample_features], dtype=float)
    X_sample_processed = preprocessing_func(X_sample)
    proba = trained_model.predict_proba(X_sample_processed)[:, 1]
    return proba[0]
````

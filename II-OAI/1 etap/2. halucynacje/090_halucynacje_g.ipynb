## Wstęp

Modele językowe pomagają nam w codziennych zadaniach, takich jak poprawianie tekstów, pisanie kodu czy odpowiadanie na pytania. 
Są one również coraz częściej wykorzystywane w takich dziedzinach jak medycyna czy edukacja.

Jednak skąd możemy wiedzieć, czy wygenerowane przez nie odpowiedzi są poprawne? Modele językowe nie zawsze posiadają pełną wiedzę na zadany temat, a mimo to mogą formułować odpowiedzi, które brzmią wiarygodnie, lecz w rzeczywistości wprowadzają w błąd. Takie niepoprawne odpowiedzi nazywamy halucynacjami.

## Zadanie

W tym zadaniu zmierzysz się z wykrywaniem halucynacji w odpowiedziach na pytania faktograficzne generowane przez duże modele językowe (LLM).
Przeanalizujesz zbiór danych, który pomoże w ocenie, czy odpowiedzi generowane przez model językowy są faktycznie poprawne, czy zawierają halucynacje.

### Kryterium Oceny

ROC AUC (ang. *Receiver Operating Characteristic Area Under Curve*) to miara jakości klasyfikatora binarnego. Pokazuje zdolność modelu do odróżniania między dwiema klasami - tutaj halucynacją (false) i poprawną odpowiedzią (true). 

- **ROC (Receiver Operating Characteristic)**: Wykres pokazujący zależność między *True Positive Rate* (czułość) a *False Positive Rate* (1-specyficzność) przy różnych progach decyzyjnych.
- **AUC (Area Under Curve)**: Pole pod wykresem ROC, które przyjmuje wartości od 0 do 1:
  - **1.0**: Model perfekcyjny.
  - **0.5**: Model losowy (brak zdolności do odróżniania klas).

Im wyższa wartość AUC, tym lepiej model radzi sobie z klasyfikacją.

Za to zadanie możesz zdobyć pomiędzy 0 a 100 punktów. Wynik będzie skalowany liniowo w zależności od wartości ROC AUC:

- **ROC AUC ≤ 0.7**: 0 punktów.
- **ROC AUC ≥ 0.82**: 100 punktów.
- **Wartości pomiędzy 0.7 a 0.82**: skalowane liniowo.

Wzór na wynik:  
$$
\text{Punkty} = 
\begin{cases} 
0 & \text{dla } \text{ROC AUC} \leq 0.7 \\
100 \times \frac{\text{ROC AUC} - 0.7}{0.82 - 0.7} & \text{dla } 0.7 < \text{ROC AUC} < 0.82 \\
100 & \text{dla } \text{ROC AUC} \geq 0.82
\end{cases}
$$

## Kod z Kryterium Oceniającym

Kod, zbliżony do poniższego, będzie używany do oceny rozwiązania na zbiorze testowym.

````python
######################### NIE ZMIENIAJ TEJ KOMÓRKI PODCZAS WYSYŁANIA ##########################

def compute_score(roc_auc: float) -> float:
    """
    Oblicza wynik punktowy na podstawie wartości ROC AUC.

    :param roc_auc: Wartość float w zakresie [0.0, 1.0]
    :return: Wynik punktowy zgodny z określoną funkcją
    """
    if roc_auc <= 0.7:
        return 0
    elif 0.7 < roc_auc < 0.82:
        return int(round(100 * (roc_auc - 0.7) / (0.82 - 0.7)))
    else:
        return 100


def evaluate_algorithm(dataset, algorithm, verbose=False):
    """
    Ewaluacja algorytmu wykrywania halucynacji na podanym zbiorze danych.

    Parametry
    ----------
    dataset : list
        Oznaczony zbiór danych, gdzie każdy element to słownik zawierający klucz 'is_correct'.
    algorithm : callable
        Funkcja, która przyjmuje pojedynczy przykład (słownik) i zwraca prawdopodobieństwo halucynacji.
    verbose : bool
        Jeśli True, wypisuje dodatkowe informacje dla każdego przykładu oraz podsumowanie.

    Zwraca
    -------
    roc_auc : float
        Wartość pola pod krzywą ROC (ROC AUC) dla predykcji.
    """
    predicted_ys = [] # Lista przechowująca przewidywane prawdopodobieństwa halucynacji

    for i, entry in enumerate(dataset):
        # Tworzenie kopii próbki i usunięcie etykiety, aby uzyskać dane wejściowe bez oznaczeń
        sample_unlabeled = dict(entry)
        sample_unlabeled.pop('is_correct', None)

        try:
            # Przewidywanie prawdopodobieństwa dla pojedynczej próbki
            pred_prob = algorithm(sample_unlabeled)
            predicted_ys.append(pred_prob)

        except Exception as e:
            # Jeśli wystąpi błąd, domyślnie ustawiamy prawdopodobieństwo na 0.5
            predicted_ys.append(0.5)
            if verbose:
                print(f"Sample {i} => Error: {e}")

    predicted_ys = np.array(predicted_ys, dtype=np.float32)
    ys = []
    for entry in dataset:
        ys.append(1 if entry.get('is_correct') else 0)
    ys = np.array(ys, dtype=np.int32)
    
    # Obliczenie metryki ROC AUC
    roc_auc = roc_auc_score(ys, predicted_ys)

    # Obliczenie końcowego wyniku na podstawie ROC AUC
    points = compute_score(roc_auc)

    if verbose:
        print(f"\nLiczba próbek: {len(dataset)}")
        print(f"ROC AUC: {roc_auc:.4f}")
        print(f"Wynik punktowy: {points}")

    return points
````

# Twoje Rozwiązanie
W tej sekcji należy umieścić Twoje rozwiązanie. Wprowadzaj zmiany wyłącznie tutaj!

````python
# Feature extraction and model training
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingClassifier
import re
import numpy as np
import xgboost as xgb
from sklearn.metrics import roc_auc_score
from collections import Counter

def extract_features(sample):
    features = {}
    
    # Basic text features
    features['answer_length'] = len(sample['answer'])
    features['question_length'] = len(sample['question'])
    features['answer_to_question_ratio'] = features['answer_length'] / max(1, features['question_length'])
    
    # Token probability features from main answer
    if 'tokens' in sample:
        tokens = sample['tokens']
        tokens_count = len(tokens)
        features['tokens_count'] = tokens_count
        
        # Analyze token text patterns
        token_text = ' '.join(tokens).lower()
        features['token_contains_maybe'] = 1 if any(word in token_text for word in ['maybe', 'perhaps', 'possibly']) else 0
        features['token_contains_numbers'] = sum(c.isdigit() for c in token_text) / max(1, len(token_text))
    
    # Supporting answers and probabilities features
    if 'supporting_probabilities' in sample and len(sample['supporting_probabilities']) > 0:
        all_probs = []
        low_prob_segments = []
        
        for probs in sample['supporting_probabilities']:
            all_probs.extend(probs)
            
            # Find segments of consecutive low probabilities
            for i in range(len(probs) - 2):
                if probs[i] < 0.5 and probs[i+1] < 0.5:
                    low_prob_segments.append((probs[i] + probs[i+1]) / 2)
            
        if all_probs:
            # Basic probability statistics
            features['supporting_prob_mean'] = np.mean(all_probs)
            features['supporting_prob_std'] = np.std(all_probs)
            features['supporting_prob_min'] = np.min(all_probs)
            features['supporting_prob_max'] = np.max(all_probs)
            features['supporting_prob_q10'] = np.percentile(all_probs, 10)
            features['supporting_prob_q25'] = np.percentile(all_probs, 25)
            features['supporting_prob_q50'] = np.median(all_probs)
            features['supporting_prob_q75'] = np.percentile(all_probs, 75)
            features['supporting_prob_q90'] = np.percentile(all_probs, 90)
            features['supporting_prob_range'] = features['supporting_prob_max'] - features['supporting_prob_min']
            features['supporting_prob_skew'] = np.mean([(p - features['supporting_prob_mean'])**3 for p in all_probs]) / max(0.001, np.std(all_probs)**3)
            
            # Distribution shape features
            below_threshold_probs = [p for p in all_probs if p < 0.5]
            features['supporting_low_prob_ratio'] = len(below_threshold_probs) / len(all_probs) if all_probs else 0
            
            # Very low probability features - potential red flags
            very_low_probs = [p for p in all_probs if p < 0.2]
            features['supporting_very_low_count'] = len(very_low_probs)
            features['supporting_very_low_ratio'] = len(very_low_probs) / len(all_probs) if all_probs else 0
            
            # Low probability segments (consecutive low probs indicate uncertainty)
            if low_prob_segments:
                features['low_prob_segments_count'] = len(low_prob_segments)
                features['low_prob_segments_mean'] = np.mean(low_prob_segments)
            else:
                features['low_prob_segments_count'] = 0
                features['low_prob_segments_mean'] = 1.0
        
        # Calculate variance across different supporting answers
        mean_probs = [np.mean(probs) for probs in sample['supporting_probabilities']]
        features['supporting_answer_mean_var'] = np.var(mean_probs) if len(mean_probs) > 1 else 0
        
        # Calculate statistics for each supporting answer separately
        for i, probs in enumerate(sample['supporting_probabilities']):
            if probs:
                features[f'supporting_{i}_prob_mean'] = np.mean(probs)
                features[f'supporting_{i}_prob_std'] = np.std(probs)
                features[f'supporting_{i}_prob_min'] = np.min(probs)
                features[f'supporting_{i}_prob_max'] = np.max(probs)
                features[f'supporting_{i}_prob_q10'] = np.percentile(probs, 10) if len(probs) >= 10 else np.min(probs)
                
                # Probability drops (sudden drops can indicate uncertainty)
                drops = [probs[i] - probs[i+1] for i in range(len(probs)-1) if probs[i] - probs[i+1] > 0.3]
                features[f'supporting_{i}_large_drops'] = len(drops)
                features[f'supporting_{i}_max_drop'] = max(drops) if drops else 0
                
                # Particularly low probability scores
                low_probs = [p for p in probs if p < 0.5]
                if low_probs:
                    features[f'supporting_{i}_low_prob_count'] = len(low_probs)
                    features[f'supporting_{i}_low_prob_mean'] = np.mean(low_probs)
                    features[f'supporting_{i}_low_prob_ratio'] = len(low_probs) / len(probs)
                else:
                    features[f'supporting_{i}_low_prob_count'] = 0
                    features[f'supporting_{i}_low_prob_mean'] = 1.0
                    features[f'supporting_{i}_low_prob_ratio'] = 0.0
    
    # Content overlap and semantic similarity between main answer and supporting answers
    if 'supporting_answers' in sample and 'answer' in sample:
        main_ans = sample['answer'].lower()
        overlap_scores = []
        
        # For tracking entities in answers
        main_entities = set()
        supporting_entities = []
        entity_patterns = r'\b([A-Z][a-z]+(?:\s+[A-Z][a-z]+)*)\b'
        
        # Extract potential entities from main answer
        for match in re.finditer(entity_patterns, sample['answer']):
            main_entities.add(match.group(1).lower())
        
        for supp_ans in sample['supporting_answers']:
            # Remove XML tags if present
            clean_supp = re.sub(r'<.*?>', '', supp_ans).lower()
            
            # Extract entities from supporting answer
            supp_entities = set()
            for match in re.finditer(entity_patterns, supp_ans):
                supp_entities.add(match.group(1).lower())
            
            supporting_entities.append(supp_entities)
            
            # Count common words (Jaccard similarity)
            main_words = set(main_ans.split())
            supp_words = set(clean_supp.split())
            
            if len(main_words) > 0 and len(supp_words) > 0:
                overlap = len(main_words.intersection(supp_words)) / len(main_words.union(supp_words))
                overlap_scores.append(overlap)
        
        if overlap_scores:
            features['mean_overlap'] = np.mean(overlap_scores)
            features['min_overlap'] = np.min(overlap_scores)
            features['max_overlap'] = np.max(overlap_scores)
            features['std_overlap'] = np.std(overlap_scores)
            
        # Entity agreement across answers
        all_supporting_entities = set().union(*supporting_entities) if supporting_entities else set()
        features['entity_agreement'] = len(main_entities.intersection(all_supporting_entities)) / max(1, len(main_entities.union(all_supporting_entities)))
    
    # Look for uncertainty markers in the answer
    uncertainty_words = ['might', 'maybe', 'could', 'possibly', 'perhaps', 'probably', 'likely', 'unlikely', 
                         'seems', 'appears', 'believe', 'think', 'assume', 'approximately', 'about', 'around']
    
    if 'answer' in sample:
        lower_ans = sample['answer'].lower()
        ans_words = lower_ans.split()
        
        # Count uncertainty markers
        uncertainty_count = sum(1 for word in uncertainty_words if word in ans_words)
        features['uncertainty_marker_count'] = uncertainty_count
        features['uncertainty_ratio'] = uncertainty_count / max(1, len(ans_words))
        
        # Check for data patterns (like dates, numbers)
        features['contains_year'] = 1 if re.search(r'\b(19|20)\d{2}\b', sample['answer']) else 0
        features['contains_number'] = 1 if re.search(r'\b\d+\b', sample['answer']) else 0
        features['numeric_density'] = len(re.findall(r'\d', sample['answer'])) / max(1, len(sample['answer']))
        
        # Sentence complexity as a potential indicator
        sentences = re.split(r'[.!?]', sample['answer'])
        if sentences:
            avg_sentence_length = np.mean([len(s.split()) for s in sentences if s.strip()])
            features['avg_sentence_length'] = avg_sentence_length
    
    # Extract answer patterns based on supporting answers
    if 'supporting_answers' in sample:
        # Check for consistency in supporting answers
        extracted_answers = []
        for supp_ans in sample['supporting_answers']:
            match = re.search(r'<answer>(.*?)</answer>', supp_ans)
            if match:
                extracted_answers.append(match.group(1).strip().lower())
        
        if extracted_answers:
            # Analyze agreement between extracted answers
            counter = Counter(extracted_answers)
            most_common_answer, most_common_count = counter.most_common(1)[0]
            
            features['extracted_answer_count'] = len(extracted_answers)
            features['unique_extracted_answer_count'] = len(counter)
            features['answer_agreement_ratio'] = most_common_count / max(1, len(extracted_answers))
            features['answer_diversity_index'] = sum((count/len(extracted_answers))**2 for count in counter.values())
            
            # Check if main answer contains the most common extracted answer
            features['contains_most_common'] = 1 if most_common_answer in sample['answer'].lower() else 0
            
            # Check for exact vs partial matches in the answer
            exact_match_count = sum(1 for ans in extracted_answers if ans in sample['answer'].lower())
            features['exact_match_ratio'] = exact_match_count / max(1, len(extracted_answers))
            
            # Check if answer tags are consistent with the question subject
            if 'question' in sample:
                question_lower = sample['question'].lower()
                question_words = set(question_lower.split())
                
                # Check if question words appear in extracted answers
                common_words = []
                for ans in extracted_answers:
                    ans_words = set(ans.split())
                    common = question_words.intersection(ans_words)
                    common_words.append(len(common))
                
                features['question_answer_word_overlap_mean'] = np.mean(common_words) if common_words else 0
                features['question_answer_word_overlap_std'] = np.std(common_words) if common_words else 0
    
    return features

# Train the hallucination detection model
def train_model(train_data):
    X_train = []
    y_train = []
    
    # Extract features from training data
    features_list = []
    for entry in train_data:
        features = extract_features(entry)
        features_list.append(features)
        X_train.append(list(features.values()))
        y_train.append(1 if entry.get('is_correct') else 0)
    
    # Get feature names for later use
    feature_names = list(features_list[0].keys())
    
    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    
    # Define base models
    base_model1 = xgb.XGBClassifier(
        n_estimators=20,
        learning_rate=0.05,
        max_depth=4,
        min_child_weight=2,
        subsample=0.5,
        colsample_bytree=0.8,
        gamma=0.1,
        objective='binary:logistic',
        use_label_encoder=False,
        eval_metric='auc',
        random_state=42
    )
    
    base_model2 = xgb.XGBClassifier(
        n_estimators=50,
        learning_rate=0.03,
        max_depth=3,
        min_child_weight=3,
        subsample=0.3,
        colsample_bytree=0.7,
        gamma=0.2,
        objective='binary:logistic',
        use_label_encoder=False,
        eval_metric='auc',
        random_state=24
    )
    
    # Create ensemble model
    model = VotingClassifier(
        estimators=[
            ('xgb1', base_model1),
            ('xgb2', base_model2)
        ],
        voting='soft'
    )
    
    # Fit the model
    model.fit(X_train_scaled, y_train)
    
    return model, scaler, feature_names

# Global variables to store the model, scaler and feature names
model, scaler, feature_names = train_model(train)
# Main prediction function
def predict_hallucinations(sample):
    global model, scaler, feature_names
    
    
    # Extract features from the input sample
    features = extract_features(sample)
    
    # Ensure feature vector has the same structure as training data
    feature_vector = []
    for name in feature_names:
        feature_vector.append(features.get(name, 0))
    
    # Scale features
    feature_vector_scaled = scaler.transform([feature_vector])
    
    # Predict probability
    prediction = model.predict_proba(feature_vector_scaled)[0, 1]
    
    # Return probability that the answer is correct (not a hallucination)
    return prediction
````

## Wstęp

Modele językowe pomagają nam w codziennych zadaniach, takich jak poprawianie tekstów, pisanie kodu czy odpowiadanie na pytania.
Są one również coraz częściej wykorzystywane w takich dziedzinach jak medycyna czy edukacja.

Jednak skąd możemy wiedzieć, czy wygenerowane przez nie odpowiedzi są poprawne? Modele językowe nie zawsze posiadają pełną wiedzę na zadany temat, a mimo to mogą formułować odpowiedzi, które brzmią wiarygodnie, lecz w rzeczywistości wprowadzają w błąd. Takie niepoprawne odpowiedzi nazywamy halucynacjami.

## Zadanie

W tym zadaniu zmierzysz się z wykrywaniem halucynacji w odpowiedziach na pytania faktograficzne generowane przez duże modele językowe (LLM).
Przeanalizujesz zbiór danych, który pomoże w ocenie, czy odpowiedzi generowane przez model językowy są faktycznie poprawne, czy zawierają halucynacje.

Każdy przykład w zbiorze danych zawiera:

- **Pytanie** np. "Jaka jest główna odpowiedzialność Departamentu Obrony USA?"
- **Odpowiedź modelu językowego** np. "Główną odpowiedzialnością jest obrona kraju."
- **Tokeny** związane z generacją odpowiedzi.

### Kryterium Oceny

ROC AUC (ang. *Receiver Operating Characteristic Area Under Curve*) to miara jakości klasyfikatora binarnego. Pokazuje zdolność modelu do odróżniania między dwiema klasami - tutaj halucynacją (false) i poprawną odpowiedzią (true).

- **ROC (Receiver Operating Characteristic)**: Wykres pokazujący zależność między *True Positive Rate* (czułość) a *False Positive Rate* (1-specyficzność) przy różnych progach decyzyjnych.
- **AUC (Area Under Curve)**: Pole pod wykresem ROC, które przyjmuje wartości od 0 do 1:
  - **1.0**: Model perfekcyjny.
  - **0.5**: Model losowy (brak zdolności do odróżniania klas).

Im wyższa wartość AUC, tym lepiej model radzi sobie z klasyfikacją.

Za to zadanie możesz zdobyć pomiędzy 0 a 100 punktów. Wynik będzie skalowany liniowo w zależności od wartości ROC AUC:

- **ROC AUC ≤ 0.7**: 0 punktów.
- **ROC AUC ≥ 0.82**: 100 punktów.
- **Wartości pomiędzy 0.7 a 0.82**: skalowane liniowo.

Wzór na wynik:  
$$
\text{Punkty} =
\begin{cases}
0 & \text{dla } \text{ROC AUC} \leq 0.7 \\
100 \times \frac{\text{ROC AUC} - 0.7}{0.82 - 0.7} & \text{dla } 0.7 < \text{ROC AUC} < 0.82 \\
100 & \text{dla } \text{ROC AUC} \geq 0.82
\end{cases}
$$

## Kod z Kryterium Oceniającym

Kod, zbliżony do poniższego, będzie używany do oceny rozwiązania na zbiorze testowym.

````python
######################### NIE ZMIENIAJ TEJ KOMÓRKI PODCZAS WYSYŁANIA ##########################

def compute_score(roc_auc: float) -> float:
    """
    Oblicza wynik punktowy na podstawie wartości ROC AUC.

    :param roc_auc: Wartość float w zakresie [0.0, 1.0]
    :return: Wynik punktowy zgodny z określoną funkcją
    """
    if roc_auc <= 0.7:
        return 0
    elif 0.7 < roc_auc < 0.82:
        return int(round(100 * (roc_auc - 0.7) / (0.82 - 0.7)))
    else:
        return 100


def evaluate_algorithm(dataset, algorithm, verbose=False):
    """
    Ewaluacja algorytmu wykrywania halucynacji na podanym zbiorze danych.

    Parametry
    ----------
    dataset : list
        Oznaczony zbiór danych, gdzie każdy element to słownik zawierający klucz 'is_correct'.
    algorithm : callable
        Funkcja, która przyjmuje pojedynczy przykład (słownik) i zwraca prawdopodobieństwo halucynacji.
    verbose : bool
        Jeśli True, wypisuje dodatkowe informacje dla każdego przykładu oraz podsumowanie.

    Zwraca
    -------
    roc_auc : float
        Wartość pola pod krzywą ROC (ROC AUC) dla predykcji.
    """
    predicted_ys = [] # Lista przechowująca przewidywane prawdopodobieństwa halucynacji

    for i, entry in enumerate(dataset):
        # Tworzenie kopii próbki i usunięcie etykiety, aby uzyskać dane wejściowe bez oznaczeń
        sample_unlabeled = dict(entry)
        sample_unlabeled.pop('is_correct', None)

        try:
            # Przewidywanie prawdopodobieństwa dla pojedynczej próbki
            pred_prob = algorithm(sample_unlabeled)
            predicted_ys.append(pred_prob)

        except Exception as e:
            # Jeśli wystąpi błąd, domyślnie ustawiamy prawdopodobieństwo na 0.5
            predicted_ys.append(0.5)
            if verbose:
                print(f"Sample {i} => Error: {e}")

    predicted_ys = np.array(predicted_ys, dtype=np.float32)
    ys = []
    for entry in dataset:
        ys.append(1 if entry.get('is_correct') else 0)
    ys = np.array(ys, dtype=np.int32)

    # Obliczenie metryki ROC AUC
    roc_auc = roc_auc_score(ys, predicted_ys)

    # Obliczenie końcowego wyniku na podstawie ROC AUC
    points = compute_score(roc_auc)

    if verbose:
        print(f"\nLiczba próbek: {len(dataset)}")
        print(f"ROC AUC: {roc_auc:.4f}")
        print(f"Wynik punktowy: {points}")

    return points
````

# Twoje Rozwiązanie
W tej sekcji należy umieścić Twoje rozwiązanie. Wprowadzaj zmiany wyłącznie tutaj!

````python
# Globalne zmienne: ensemble_model, lista nazw cech, TF-IDF vectorizer oraz scaler
ensemble_model = None
feature_names = None
global_tfidf = None
global_svd = None
scaler = None

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
import numpy as np
from sklearn.preprocessing import StandardScaler

def get_supporting_marked(sample):
    supporting_marked = []
    supporting_probs = []
    for i, suptoks in enumerate(sample.get("supporting_tokens")):
        otag = suptoks.index('<answer>') if '<answer>' in suptoks else len(suptoks)
        ctag = suptoks.index('</answer>') if '</answer>' in suptoks else len(suptoks)
        supporting_marked += [suptoks[otag+1:ctag]]
        supporting_probs += [sample['supporting_probabilities'][i][otag+1:ctag]]
    return supporting_marked, supporting_probs


def get_sup_marked_sim(sample):
    marked, probs = get_supporting_marked(sample)
    count = 0
    for i, m in enumerate(marked):
        sim = 0
        for j, t in enumerate(m):
            sim += probs[i][j] if t in sample['answer'] else 0
        count += sim / (len(m) or 1)

    return count

def extract_features(sample):
    """
    Ekstrakcja cech z pojedynczego przykładu.

    Wyciągamy:
      - Podstawowe cechy długościowe (liczba słów w pytaniu i odpowiedzi, różnica, stosunek).
      - Statystyki dla wspierających odpowiedzi (średnia, std długości).
      - Statystyki z supporting_probabilities.
      - Cechy oparte na TF-IDF: cosine similarity między pytaniem a odpowiedzią oraz między odpowiedzią a wspierającymi.
      - Cechy tokenowe.
    """
    features = {}

    # Podstawowe cechy tekstowe
    question = sample.get("question", "")
    answer = sample.get("answer", "")
    question_wc = len(question.split())
    answer_wc = len(answer.split())
    features["question_wc"] = question_wc
    features["answer_wc"] = answer_wc
    features["ratio"] = answer_wc / question_wc if question_wc > 0 else 0
    features["diff_answer_question_wc"] = answer_wc - question_wc

    # Cechy wspierających odpowiedzi (supporting_answers)
    supporting_answers = sample.get("supporting_answers", [])
    supp_word_counts = [len(ans.split()) for ans in supporting_answers]
    if supp_word_counts:
        avg_supp_wc = np.mean(supp_word_counts)
        std_supp_wc = np.std(supp_word_counts)
    else:
        avg_supp_wc = 0
        std_supp_wc = 0
    features["avg_supp_wc"] = avg_supp_wc
    features["std_supp_wc"] = std_supp_wc
    features["diff_answer_avg_supp_wc"] = answer_wc - avg_supp_wc
    features["ratio_answer_avg_supp_wc"] = answer_wc / avg_supp_wc if avg_supp_wc > 0 else 0

    # Statystyki z supporting_probabilities
    supporting_probs = sample.get("supporting_probabilities", [])
    avg_probs = [np.mean(prob_list) if prob_list else 0 for prob_list in supporting_probs]
    if avg_probs:
        features["avg_supp_prob"] = np.mean(avg_probs)
        features["std_supp_prob"] = np.std(avg_probs)
        features["max_supp_prob"] = np.max(avg_probs)
        features["min_supp_prob"] = np.min(avg_probs)
    else:
        features["avg_supp_prob"] = 0
        features["std_supp_prob"] = 0
        features["max_supp_prob"] = 0
        features["min_supp_prob"] = 0
    for i in range(4):
        features[f"supp_prob_{i}"] = avg_probs[i] if i < len(avg_probs) else 0

    # Cechy oparte na TF-IDF i cosine similarity
    if global_tfidf is not None:
        q_vec = global_svd.transform(global_tfidf.transform([question]))
        a_vec = global_svd.transform(global_tfidf.transform([answer]))
        features["cosine_qa"] = cosine_similarity(q_vec, a_vec)[0][0]

        cosine_vals = []
        for supp in supporting_answers:
            supp_vec = global_svd.transform(global_tfidf.transform([supp]))
            cosine_vals.append(cosine_similarity(a_vec, supp_vec)[0][0])
        features["cosine_as"] = np.mean(cosine_vals) if cosine_vals else 0
        features["max_cosine_as"] = np.max(cosine_vals) if cosine_vals else 0
    else:
        features["cosine_qa"] = 0
        features["cosine_as"] = 0
        features["max_cosine_as"] = 0

    # Cechy tokenowe
    main_tokens = sample.get("tokens", [])
    features["main_tokens_count"] = len(main_tokens)
    supp_tokens_list = sample.get("supporting_tokens", [])
    supp_tokens_counts = [len(tokens) for tokens in supp_tokens_list]
    if supp_tokens_counts:
        features["avg_supp_tokens_count"] = np.mean(supp_tokens_counts)
        features["std_supp_tokens_count"] = np.std(supp_tokens_counts)
        features["max_supp_tokens_count"] = np.max(supp_tokens_counts)
        features["min_supp_tokens_count"] = np.min(supp_tokens_counts)
    else:
        features["avg_supp_tokens_count"] = 0
        features["std_supp_tokens_count"] = 0
        features["max_supp_tokens_count"] = 0
        features["min_supp_tokens_count"] = 0

    features['suppporting_marked'] = get_sup_marked_sim(sample)

    return features

def train_stacking_fast(train_data):
    """
    Trenuje uproszczony stacking ensemble na danych treningowych.

    Używamy:
      - TF-IDF vectorizera z mniejszą liczbą cech (max_features=500).
      - Standaryzacji cech.
      - Bazowych modeli: XGBoost, Random Forest oraz regresji logistycznej.
      - Finalnego meta modelu (regresja logistyczna).
    """
    global global_tfidf, global_svd, scaler
    # Przygotowanie TF-IDF vectorizera – mniejszy rozmiar słownika dla szybszego przetwarzania
    documents = []
    for sample in train_data:
        documents.append(sample.get("question", ""))
        documents.append(sample.get("answer", ""))
        documents.extend(sample.get("supporting_answers", []))
    global_tfidf = TfidfVectorizer(max_features=2048)
    tfidf = global_tfidf.fit_transform(documents)
    global_svd = TruncatedSVD(n_components=50)
    global_svd.fit(tfidf)
    X = []
    y = []
    for sample in train_data:
        feats = extract_features(sample)
        X.append(list(feats.values()))
        y.append(1 if sample.get("is_correct") else 0)
    feat_names = list(extract_features(train_data[0]).keys())
    X = np.array(X)
    y = np.array(y)

    # Skalowanie cech
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Definicja bazowych modeli
    estimators = [
        ('xgb', xgb.XGBClassifier(
            use_label_encoder=False,
            eval_metric='logloss',
            max_depth=15,
            n_estimators=500,  # zmniejszona liczba estymatorów
            learning_rate=0.01,
            subsample=0.8,
            colsample_bytree=0.8,
            verbosity=1,
            verbose=1,
            scale_pos_weight =  1 / (2015 / 2967),

        )),
        ('rf', RandomForestClassifier(n_estimators=100, max_depth=15, random_state=42)),
        ('lr', LogisticRegression(max_iter=500, solver='liblinear'))
    ]

    # Uproszczony stacking ensemble bez kosztownych operacji
    stacking = StackingClassifier(
        estimators=estimators,
        final_estimator=LogisticRegression(max_iter=500, solver='liblinear'),
        cv=2,
        n_jobs=-1,
        verbose=1
    )
    stacking.fit(X_scaled, y)
    return stacking, feat_names

def predict_hallucinations(sample):
    """
    Dla danego przykładu zwraca prawdopodobieństwo, że odpowiedź jest poprawna.

    Przy pierwszym wywołaniu model jest trenowany na danych treningowych.
    """
    global ensemble_model, feature_names, scaler
    if ensemble_model is None:
        try:
            ensemble_model, feature_names = train_stacking_fast(train)
        except Exception as e:
            print("Błąd podczas trenowania ensemble:", e)
            return 0.5
    feats = extract_features(sample)
    X_sample = np.array([feats[feat] for feat in feature_names]).reshape(1, -1)
    X_sample_scaled = scaler.transform(X_sample)
    try:
        prob = ensemble_model.predict_proba(X_sample_scaled)[0, 1]
    except Exception as e:
        print("Błąd podczas predykcji:", e)
        prob = 0.5
    return float(prob)
````
